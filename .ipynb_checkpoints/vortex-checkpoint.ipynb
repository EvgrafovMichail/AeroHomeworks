# Вихри. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = 50

x_beg, x_end = -2, 2
y_beg, y_end = -1.5, 1.5
x0, y0 = 0, 0

strenght = 5
intensity = -1

vortex_num = 5
step = 1

lenght = 12.4
```

```python
def get_koeff(x_beg, x_end, y_beg, y_end):
    
    k = 0
    
    if abs(x_beg - x_end) != 0:
        k = abs(y_beg - y_end) / abs(x_beg - x_end)
        
    return k
```

```python
def  get_grid(x_beg, x_end, y_beg, y_end, N):
    
    x = np.linspace(x_beg, x_end, N)
    y = np.linspace(y_beg, y_end, N)
    X, Y = np.meshgrid(x, y)
    
    return X, Y
```

```python
def get_velocity_vortex(strenght, x0, y0, X, Y):
    
    U = strenght / 2 / np.pi * (Y - y0) / ((X - x0) ** 2 + (Y - y0) ** 2)
    V = - strenght / 2 / np.pi * (X - x0) / ((X - x0) ** 2 + (Y - y0) ** 2)
    
    return U, V
```

```python
def get_velocity_sink(sigma, x0, y0, X, Y):
    
    U = sigma / 2 / np.pi * (X - x0) / ((X - x0) ** 2 + (Y - y0) ** 2)
    V = sigma / 2 / np.pi * (Y - y0) / ((X - x0) ** 2 + (Y - y0) ** 2)
    
    return U, V
```

```python
def get_velocity_infinit_vortex(gamma, a, X, Y):
    
    U = gamma / 2 / a * np.sinh(2 * np.pi * Y / a)
    V = - gamma / 2 / a * np.sin(2 * np.pi * X / a)
    den = np.cosh(2 * np.pi * Y / a) - np.cos(2 * np.pi * X / a)
    
    return U / den, V / den
```

```python
def get_multi_velocity(func, strenght, X0, Y0, X, Y):
    
    size = min(len(X0), len(Y0))
    
    U = [func(strenght, X0[i], Y0[i], X, Y)[0] for i in range(size)]
    V = [func(strenght, X0[i], Y0[i], X, Y)[1] for i in range(size)]
    
    return np.sum(U, axis=0), np.sum(V, axis=0)
```

## Вихрь и сток. 

```python
k = get_koeff(x_beg, x_end, y_beg, y_end)

fig, ax = plt.subplots(figsize=(lenght, k * lenght))

X, Y = get_grid(x_beg, x_end, y_beg, y_end, N)

u_vortex, v_vortex = get_velocity_vortex(strenght, x0, y0, X, Y)
u_sink, v_sink = get_velocity_sink(intensity, x0, y0, X, Y)
u, v = u_vortex + u_sink, v_vortex + v_sink

ax.set_title('Vortex', fontsize=20, fontweight='bold')

ax.streamplot(X, Y, u, v, color='b', density=2.5, linewidth=1, 
              arrowsize=1, arrowstyle='->')
ax.scatter(x0, y0, color='r', marker='o')

ax.set_xlim(x_beg, x_end)
ax.set_ylim(y_beg, y_end)

ax.grid();
```

## Система вихрей и стоков.

```python
k = get_koeff(x_beg, x_end, y_beg, y_end)

fig, ax = plt.subplots(figsize=(lenght, k * lenght))

X, Y = get_grid(x_beg, x_end, y_beg, y_end, N)
X0 = np.linspace(x_beg, x_end, vortex_num)
Y0 = np.zeros(shape=vortex_num)

u_vortex, v_vortex = get_multi_velocity(get_velocity_vortex, strenght, X0, Y0, X, Y)
u_sink, v_sink = get_multi_velocity(get_velocity_sink, intensity, X0, Y0, X, Y)
u, v = u_vortex + u_sink, v_vortex + v_sink

ax.set_title('Row of Vortices', fontsize=20, fontweight='bold')

ax.streamplot(X, Y, u, v, color='b', density=2.5, linewidth=1, 
              arrowsize=1, arrowstyle='->')
ax.scatter(X0, Y0, color='r', marker='o')

ax.set_xlim(x_beg, x_end)
ax.set_ylim(y_beg, y_end)

ax.grid();
```

## Бесконечная цепочка вихрей.

```python
k = get_koeff(x_beg, x_end, y_beg, y_end)

fig, ax = plt.subplots(figsize=(lenght, k * lenght))

X, Y = get_grid(x_beg, x_end, y_beg, y_end, N)

u, v = get_velocity_infinit_vortex(strenght, step, X, Y)

ax.set_title("Vortex's infinit row" , fontsize=20, fontweight='bold')

ax.streamplot(X, Y, u, v, color='b', density=2.5, linewidth=1, 
              arrowsize=1, arrowstyle='->')

ax.set_xlim(x_beg, x_end)
ax.set_ylim(y_beg, y_end)

ax.grid();
```
