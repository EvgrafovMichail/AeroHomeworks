## Используемые библиотеки: 

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Константы:

```python
lenght = 12.8                                   # длина изображения в дюймах
width = 9.6                                     # ширина изображения в дюймах

x_start, x_end = -1, 2                          # границы x расчётной сетки
y_start, y_end = -0.5, 0.5                      # границы y расчётной сетки
N = 51                                          # количество узлов в сетке

u_inf = 1                                       # скорость набегающего потока

X0 = np.loadtxt('NACA0012_x.txt')               # абсциссы источников
Y0 = np.loadtxt('NACA0012_y.txt')               # ординаты источников
intensity = np.loadtxt('NACA0012_sigma.txt')    # интенсивности источников
```

## Вспомогательные функции:

```python
def get_min(num1, num2, num3):
    """
    Эта функция определяет минимальное из трёх чисел
    
    :param: num1 - число
    :param: num2 - число
    :param: num3 - число
    
    :return: значение минимального из трёх чисел
    
    """
    return (min(num1, min(num2, num3)))
```

```python
def get_grid(N, x_start, x_end, y_start, y_end):
    """
    Эта функция генерирует узлы расчётной сетки
    
    :param: N - количество узлов в обоих направлениях - число
    :param: x_start - нижняя граница абсциссы - число
    :param: x_end - верхняя граница абсциссы - число
    :param: y_start - нижняя граница ординаты - число
    :param: y_end - вурхняя граница ординаты - число
    
    :return: X, Y - массивы координат расчётной сетки
    
    """
    x = np.linspace(x_start, x_end, N)
    y = np.linspace(y_start, y_end, N)
    X, Y = np.meshgrid(x, y)
    
    return X, Y
```

```python
def calculate_source_velocity(sigma, x0, y0, X, Y):
    """
    Эта функция рассчитывает скорость потока, вытекающего
    из источника, в кажом узле сетки
    
    :param: sigma - интенсивность потока - число 
    :param: x0 - абсцисса источника - число
    :param: y0 - ордината источника - число
    :param: X - массив координат x расчётной сетки
    :param: Y - массив координат y расчётной сетки
    
    :return: u, v - массивы значений компонент скорости
                    в каждом узле сетки
    
    """
    u = sigma / (2 * np.pi) * (X - x0) / ((X - x0) ** 2 + (Y - y0) ** 2)
    v = sigma / (2 * np.pi) * (Y - y0) / ((X - x0) ** 2 + (Y - y0) ** 2)
    
    return u, v
```

```python
def calculate_summary_source_velocity(sigmas, X0, Y0, X, Y):
    """
    Эта функция рассчитывает суммарную скорость потока, 
    вытекающего из нескольких источников, в каждом узле сетки
    
    :param: sigmas - массив интенсивностей потоков 
    :param: X0 - массив абсцисс источника
    :param: y0 - массив ординат источника
    :param: X - массив координат x расчётной сетки
    :param: Y - массив координат y расчётной сетки
    
    :return: u, v - массивы значений компонент скорости
                    в каждом узле сетки
    
    """
    
    size = get_min(len(sigmas), len(X0), len(Y0))
    
    u = [calculate_source_velocity(sigmas[i], X0[i], Y0[i], X, Y)[0] for i in range(size)]
    v = [calculate_source_velocity(sigmas[i], X0[i], Y0[i], X, Y)[1] for i in range(size)]
    
    return np.sum(np.array(u), 0), np.sum(np.array(v), 0)
```

```python
def calculate_stream_velocity(nodes_num, u_inf, alpha):
    """
    Эта функция рассчитывает скорость набегающего потока
    в каждом усзле сетки
    
    :param: nodes_num - количество узлов сетки в одном направлении - число 
    :param: u_inf - скорость потока - число
    :param: alpha - угол набегания- число
    
    :return: u, v - массивы значений компонент скорости
                    в каждом узле сетки
    
    """
    
    u = u_inf * np.ones((nodes_num, nodes_num)) * np.cos(alpha)
    v = u_inf * np.ones((nodes_num, nodes_num)) * np.sin(alpha)
    
    return u, v
```

```python
def calculate_pressure_coefficient(U, V, u_inf):
    """
    Эта функция рассчитывает коэффициент давления
    в каждом узле сетки
    
    :param: U - массив компонент x скоростей в каждом узле 
    :param: V - массив компонент y скоростей в каждом узле
    :param: u_inf - скорость потока - число
    
    :return: cp - массив значений коэффициента давления 
                в каждом узле сетки  
    """
    
    if u_inf:
        
        return 1 - (U ** 2 + V ** 2) / u_inf ** 2
    
    return None
```

## Пункт 1. Вычисление скоростей.

```python
fig, ax = plt.subplots(figsize=(lenght, width))

X, Y = get_grid(N, x_start, x_end, y_start, y_end)

u_source, v_source = calculate_summary_source_velocity(intensity, X0, Y0, X, Y)
u_stream, v_stream = calculate_stream_velocity(N, u_inf, 0)

u, v = u_source + u_stream, v_source + v_stream

ax.set_xlim(x_start, x_end)
ax.set_ylim(y_start, y_end)
ax.set_xlabel('x', fontsize=16, fontweight='bold')
ax.set_ylabel('y', fontsize=16, fontweight='bold')
ax.set_title('Sources + Stream', fontsize=20, fontweight='bold')

ax.streamplot(X, Y, u, v, color='b', density=2.5, linewidth=1, 
              arrowsize=1, arrowstyle='->')
ax.scatter(X0, Y0, s=30, color='r', marker='o');

ax.grid()
```

## Пункт 2. Вычисление коэффициента давления. 

```python
fig, ax = plt.subplots(figsize=(lenght, width))

cp = calculate_pressure_coefficient(u, v, u_inf)
max_ind = np.argmax(cp)
i, j = np.unravel_index(max_ind, (N, N))

ax.set_xlabel('x', fontsize=16, fontweight='bold')
ax.set_ylabel('y', fontsize=16, fontweight='bold')
ax.set_xlim(x_start, x_end)
ax.set_ylim(y_start, y_end)
ax.set_title('Pressure Coefficient Distribution', fontsize=20, fontweight='bold')

contf = plt.contourf(X, Y, cp, levels=np.linspace(cp.min(), cp.max(), 100), 
                     extend='both')
cbar = plt.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks(np.linspace(round(cp.min(), 1), round(cp.max(), 1), 11))

ax.scatter(X[i][j], Y[i][j], s=30, color='r', marker='o', label='- maximum')

ax.legend();
```
